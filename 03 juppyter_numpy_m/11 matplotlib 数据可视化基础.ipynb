## matplotlib 基础
> **通常只进行简单的处理，只使用到matplotlib的一个子模块pyplot**

> **plot方法绘制图形,只是返回一个内存对象。show方法返回图形,进行渲染**

> **plot方法可以绘制多条曲线，只是在最后一步进行渲染**

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10 , 100)
```

```python
y = np.sin(x)
```

```python
plt.plot(x, y)
plt.show()
```

---
>plot绘制多条曲线，只需在最后一步进行渲染即可

> **plot传入新的参数color='red',修改颜色，传入参数linestyle,改变曲线的样式**

> **plot传入参数label的属性，最后调用legend()方法显示图示，说明图示**

>plt调用xlim方法和ylim方法或axis方法，修改坐标系

>plt调用xlabel和ylable，说明x轴或y轴的意义

>plt调用title添加标题

```python
cosy = np.cos(x)
```

```python
siny = y.copy()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy)
plt.show()
```

```python
plt.plot(x, siny, color='red')
plt.plot(x, cosy)
plt.show()
```

```python
plt.plot(x, siny, color='red')
plt.plot(x, cosy, linestyle='--')
plt.show()
```

```python
plt.plot(x, siny, color='red')
plt.plot(x, cosy, linestyle='--')
plt.xlim(-5 , 15)
plt.ylim(-2 , 2)
plt.show()
```

```python
plt.plot(x, siny, color='red')
plt.plot(x, cosy, linestyle='--')
plt.axis([-1, 11, -2 , 2])
plt.show()
```

```python
plt.plot(x, siny, color='red')
plt.plot(x, cosy, linestyle='--')
plt.xlabel("x label")
plt.ylabel("y label")
plt.show() 
```

```python
plt.plot(x, siny, color='red',label='sin(x)')
plt.plot(x, cosy, linestyle='--', label='cos(x)')
plt.xlabel("x label")
plt.ylabel("y label")
plt.legend()##还需要调用legend()方法显示图示
plt.show() 
```

```python
plt.plot(x, siny, color='red',label='sin(x)')
plt.plot(x, cosy, linestyle='--', label='cos(x)')
plt.xlabel("x label")
plt.ylabel("y label")
plt.legend()##还需要调用legend()方法显示图示
plt.title("welcome to the machine learning")
plt.show() 
```

## scatter plot 散点图
> **调用scatter方法绘制散点图**

> **方法的调用于plot折线图相同, 可以传入参数alpha表示透明度**

> **<font color=red>折线图用于绘制一维特征， 散点图用于绘制二维特征</font>**

```python
plt.scatter(x, siny)
plt.show()
```

```python
plt.scatter(x, siny, color='red',label='sin(x)')
plt.scatter(x, cosy, linestyle='--', label='cos(x)')
plt.xlabel("x label")
plt.ylabel("y label")
plt.legend()##还需要调用legend()方法显示图示
plt.show() 
```

```python
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)
plt.scatter(x, y)
plt.show()
```

```python
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)
plt.scatter(x, y, alpha = 0.5)
plt.show()
```
